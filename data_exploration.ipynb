```python
import pandas as pd
import numpy as np
```

```python
#loading batting file
file = "baseball data/baseballdatabank-2019.2/core/Batting.csv"
statdf = pd.read_csv(file)
```

```python
#cleaning up dataset to only include players with at least 200 ABs
statdf = statdf.dropna()
statdf = statdf.loc[statdf['AB']>=200, :]
statdf.head()
```

```python
#organizing dataset by player & year
dfplayer = statdf.groupby(['playerID','yearID']).sum()
dfplayer.head()
```

```python
#loading player info dataset
file = "baseball data/baseballdatabank-2019.2/core/People.csv"
persondf = pd.read_csv(file)
persondf.head()
```

```python
#grabbing birthyear from player info dataset
persondf = persondf[['playerID','birthYear']]
persondf.head()
```

```python
#checking datatype of playerID column
statdf.playerID
```

```python
#merging dataframes on playerID
maindf = pd.merge(persondf, statdf, on='playerID', how='right')
maindf.head()
```

```python
#fixing datatypes in order to do calculations
maindf['yearID'] = maindf.yearID.astype('float64')
maindf['birthYear'] = maindf.birthYear.astype('float64')
maindf = maindf.astype({'AB' : 'float64','H' : 'float64','2B' : 'float64','3B' : 'float64','HR' : 'float64',})
```

```python
#calculating age, batting average, slugging, and isolated power
maindf['age'] = maindf['yearID'] - maindf['birthYear']
maindf['SLG'] = (maindf['H'] + (2*maindf['2B']) + (3*maindf['3B']) + (4*maindf['HR']))/maindf['AB']
maindf['AVG'] = maindf['H'] / maindf['AB']
maindf['ISO'] = maindf['SLG'] - maindf['AVG']
maindf.head()
```

```python
#dropping birthyear to be replaced by age
maindf = maindf.drop(columns ='birthYear')
maindf.head()
```

```python
#saving new dataframe 
maindf.to_csv('baseball data/player_stats.csv')
```
